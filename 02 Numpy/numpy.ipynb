- Deep Copy and Shallow Copy

THEORY ---> Deep copy means separate copy and no changes on original data while Shallow copy means changes made will be reflected in the original data too. Whenever we are slicing any portion of an array they make their own space or location in RAM that's why they don't make any changes in original array. But if we copy the array and made changes in it then it will change the original array too.

```python
a = np.arange(1,21)

print(a)
```

```python
a_slice = a[:6]              #slicing a portion

print(a_slice)
```

```python
a_slice = a_slice * 10         #broadcasting in a sliced portion

print(a_slice)
```

```python
print(a)              # when we make changes in the sliced portion of an array , then it will not affect the original array , so this is called deep copy.
```

```python
b = a       #if we copy the  original array 

b[0] = 99     #now make changes it in.
```

```python
print(b)
```

```python
print(a)         #changes appears in the original array too
```

```python
a = np.arange(1,21)

print(a)
```

```python
b = a.copy()       # if we assign the copy of the original array 

b[0] = 99         # now make changes it in.
```

```python
print(b)
```

```python
print(a)       # no changes in original array because we assign copy of original array to b.
```

- Valid Soduku

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

Solution to prove a soduku a valid soduku

```python
a = np.sum(s, axis=0)   # we check whether the sum of all the elements of the each row is 45 or not (sum of all the element in each row is 45 in a valid soduku)

for i in a:
    if i != 45:
        print("this is not a valid soduku")
        break
else:
    print("currently for rows, soduku is valid")
```

```python
b = np.sum(s, axis=1)    # we check whether the sum of all the elements of the each column is 45 or not (sum of all the element in each column is 45 in a valid soduku)

for i in b:
    if i != 45:
        print("this is not a valid soduku")
        break
else:
    print("for columns also , soduku is valid")

```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n = s[i:i+3,j:j+3]
        print(n.sum())

```
