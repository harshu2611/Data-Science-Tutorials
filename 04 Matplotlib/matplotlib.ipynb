THEORY ----> Key terms you should know

1. Data Points --> single point in a dataset (both for x axis and y axis)
2. x-axis --> horizontal line where basically input data point are present
3. y-axis --> vertical line where basically output data point are present
4. figure --> the entire canvas where the graph is plot.
5. Axes/graph box --> the actual area where the graph is plotted inside the canvas.
6. Plot --> the actual line, bar, point and shape which we draw inside Axes. and we actually see it.
7. Marker --> the symbol which is used to visualize the data points. eg: dot, square, diamond, cross, plus sign etc...,
8. line style --> the type of line we want to draw between two data points. eg:- solid, dashed, dotted lines.
9. Color --> it defines the color of lines, bars, data points
10. Legend --> used to represent which line denotes what.
11. Label --> use to represent which axis denotes what.
12. Title --> use to give headline to the graph. (for what the graph is)
13. Grid --> it is cross-sections on the graph.
14. Function --> block of codes use to perform specific tasks.
15. Methods --> in advanced cases like object oriented APIs we use methods
16. Parameters --> informations given to a function to control what they do.
17. Keyword arguments --> this keyword is normally pass with vlaues. and usually this keywords are not passed.
18. Object-Oriented API --> advanced method to create plots
19. DPI --> (dot per inch) higher the dpi clearer the image.
20. Backend --> used to save the images of plot or graph.

```python
month = ['january','february','march','april','may','june']
sales = [1000,1500,1200,900,2000,1700]

plt.plot(month,sales, color='skyblue', linestyle= '-', linewidth= '2', marker= 'o', label= 'Half yearly Sales data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Record')
plt.legend()
plt.grid(color='red', linestyle=':')
plt.xlim()
plt.ylim()
plt.xticks(rotation=60)

plt.show()
```

- Bar Plot

plt.bar(x, height, color='color name', width= value, label='label name')   # for vertical bars


plt.barh(x, height, color='color name', width= value, label='label name')   # for horizontal bars

```python
product = ['A','B','C','D','E']
sales = [1000,1500,800,1200,700]

plt.bar(product,sales, color='orange', label = 'Sales 2025', width= 0.5)

plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product sales Comparison')
plt.legend()

plt.show()
```

- Pie Chart

plt.pie (values , label= label_list, color= color_list, autopct='%1.1f%%')

'% means percentage
1.1f means 1 digit before decimal and 1 digit after decimal
%% means after value put percentage (eg: 37.5% )

```python
regions = ['east','west','north','south']
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'] )
plt.title('Revenue Contribution of Each Region')

plt.show()
```

- Histogram 

plt.hist(data, bins=value, color='color name', edgecolor='color name')

```python
scores = [45,67,89,56,78,88,92,60,75,26,48,19,54,8,39,65,98,53,90,42]

plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distribution of Students')

plt.show()
```

- Subplot

1. multiple charts in one place
2. less space needed
3. quickly analyse multiple datasets together

plt.subplot(nrows, ncols, index)

```python
plt.subplot(2,2,1)
plt.plot([1,2,3,4,5,6,7,8],[1,4,9,16,25,36,49,64])


plt.subplot(2,2,4)
plt.plot([1,4,9,16,25,36,49,64],[1,2,3,4,5,6,7,8])

# similarily 2 and 3 can also be plotted
```

```python
product = ['A','B','C','D','E']
sales = [1000,1500,800,1200,700]

plt.subplot(1,2,1)   # 1 row , 2 column, 1st subplot
plt.plot(product,sales,  color='green', linestyle= '-', linewidth= '1', marker= 'o', label= 'Sales data')
plt.xlabel('product')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Record')
plt.legend()
plt.grid(color='red', linestyle=':')

plt.subplot(1,2,2)   # 1 row , 2 column, 2nd subplot
plt.bar(product,sales, color='orange', label = 'Sales 2025', width= 0.5)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product sales Comparison')
plt.legend()

plt.show()
```

Object Oriented Way

fig, ax = plt.subplot(nrows, ncols, figsize=(width, height))

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))


product = ['A','B','C','D','E']
sales = [1000,1500,800,1200,700]

ax[0].plot(product,sales, color='blue')
ax[0].set_title('line chart')

ax[1].bar(product,sales, color='green')
ax[1].set_title('bar chart')

plt.suptitle('Comparison of line and bar chart')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))

axis1 = fig.add_axes([0.1,0.1,0.7,0.7])    # [distance from left, distance from bottom, width,height]
axis1.plot(product, sales)

axis2 = fig.add_axes([0.6,0.6,0.2,0.2])    # [distance from left, distance from bottom, width,height]
axis2.plot(product, sales) 
```

- Savefig() function

savefig('filename.extension', dpi= value, bbox_inches='tight')   #filemane.extension is filename with extension like png, jpeg, pdf etc...   bbox_inches is use to fill the image in full space without any whitespace 

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y, color='blue', marker='o')
plt.title('Simple line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.savefig('line_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```
