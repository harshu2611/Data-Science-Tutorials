# Distribution Plots

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

- Histplot/distplot

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins= 20, kde=True)

plt.subplot(1,2,2)
sns.histplot(df['tip'], bins= 20, kde=True)
```

- Joint plot

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='scatter')   # scatter kind is used
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='hex')   # hex kind is used
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='kde')   # kde kind is used
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='reg',)   # reg kind is used
```

- Pair Plot

```python
sns.pairplot(df)    # it compares all numerical valued colums eg:- total_bill, tips, size
```

```python
sns.pairplot(df, hue='sex')     # hue differentiate each graph for male as well as female
```

```python
sns.pairplot(df, hue='size', palette='rainbow')   # palette is used for colors
```

- Rug plot

```python
sns.rugplot(df)   # show lines where value present for total_bill, tip and size
```

# Categorical Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

stripplot
swarmplot

- countplot

```python
sns.countplot(x=df['sex'])    # it counts the total number of male and female
```

```python
sns.countplot(x=df['sex'], hue= df['smoker'])    # it gives total number of male and female who smoke and who not smoke
```

- barplot

x-axis contain categorical valued column and y-axis has quantitative valued column.

```python
sns.barplot(x= df['sex'], y= df['total_bill'])
```


You can change the estimator object to your own function, that converts a vector to a scalar 
- Mean (default): np.mean 
- Median: np.median 
- Sum: np.sum 
- Count: len (to count the number of observations) 
- Standard Deviation: np.std 
- Minimum: np.min 
- Maximum: np.max

```python
sns.barplot(x= df['sex'], y= df['total_bill'], estimator=np.sum)    # by deafult estimator is mean
```

```python
sns.barplot(x= df['sex'], y= df['tip'], estimator=np.sum)   
```

- boxplot

```python
sns.boxplot(data=df, x='day', y='tip', palette='rainbow')
```

#### boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

- violinplot

```python
sns.violinplot(data= df, x='day', y='total_bill')
```

#### stripplot and swarmplot

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

- stripplot

```python
sns.stripplot(data= df, x='day', y='total_bill')
```

- swarmplot

```python
sns.swarmplot(data= df, x='day', y='total_bill')
```

- violinplot and swarmplot together

```python
sns.violinplot(data= df, x='day', y='tip')
sns.swarmplot(data= df, x='day', y='tip')
```

- Heatmap

If one value increases with increase in other value then it is called positive correlation , On the other hand if it decreases with increase in other value then it is called negative correlation. ( The value of correlation ranges from +1 to -1). To plot correlation into plots we use heatmaps.

```python
tipscorr = tips[['total_bill','tip','size']]    # we use only columns which have numerical values
```

```python
tipscorr
```

```python
tipscorr.corr()     # here we find correlation.
```

```python
sns.heatmap(tipscorr.corr(), annot=True)    # we plot a heatmap here.
```
