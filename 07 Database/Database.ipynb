# SQL (Structured Query Language)

- DDL (data definition language)
    - create (create new database object)
    - alter (modifying existing database objects)
    - drop (delete existing database onjects)
    - truncate (remove all rows form table)
- DML (data manipulation language)
    - insert (create new rows in tables)
    - update (modify data in tables)
    - delete (delete data from tables)
    - select (retrieve data from tables)
- DCL (data control language)
    - grant (provide access rights on database)
    - revoke (withdraw access rights on database)
- TCL ( transaction control language)
    - commit (save database changes and end transaction)
    - rollback (undo changes that are not committed and end transaction)

# SUBQUERY

- In SQL, a subquery is a query within another query. 
- It as a SELECT statement that is nested inside another SELECT, INSERT, UPDATE or DELETE statement.
- The subquery is executed first, and its result is then used as a parameter or condition for the outer query. 

- types of subqueries
    1. based on the result it returns
        - Scalar subquery
        - Row subquery
        - Table subquery
    2. based on working
        - Independent
        - correlated
