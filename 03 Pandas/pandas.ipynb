```python
df      #NaN are null or missing values
```

```python
df.isna().sum()        #it gives the total number of NaN in each columns
```

```python
df.isna().any()      #it gives true if a column has any NaN value and it gives false if a column has zero NaN values.
```

- Removing Missing Data

```python
df
```

```python
df.dropna()    # it will remove all the rows which has NaN values.
```

```python
df.dropna(thresh=3)       # if the row not have atleast 3 non NaN values , then it will be deleted (row 2 and 3 don't have 3 non NaN values so they are removed)
```

# Merging Joining and Concatenation

- Merging 2 Dataframes

```python
# DataFrame 1: Employees information
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})

```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries)       # only 3 employee ids came because they are only common employee ids.
```

```python
pd.merge(employees, salaries, on='employee_id')       # on keyword is used because we merge the dataframes on the basis of columns.
```

```python
pd.merge(employees, salaries, on='employee_id', how='inner')    # inner means we merge on the basis of common employee_id.
```

```python
pd.merge(employees, salaries, on='employee_id', how='outer')    # outer means we merge on the basis of complete column employee_id and return each employee_id and also it returns NaN at missing values.
```

```python
pd.merge(employees, salaries, on='employee_id', how='left')  # merge on the basis of left data we passed (employeed here) , so we get every data of employees here.
```

```python
pd.merge(employees, salaries, on='employee_id', how='right')  # merge on the basis of right data we passed (salaries here) , so we get every data of salaries here.
```

# GroupBy and Aggregation

- Group By

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by Category and calculate the sum of Sales

cat_sum = df.groupby('Category')['Sales'].sum()   # it returns the sum of sales for each category.

cat_sum
```

```python
# Group by Store and calculate the sum of Sales

sto_sum = df.groupby('Store')['Sales'].sum()

sto_sum
```

```python
# Group by multiple columns
# Group by Category and Store

cat_sto = df.groupby(['Category', 'Store'])['Sales'].sum()

cat_sto
```

-make a new column for episode count

```python
df.loc[2]['Title']   #observed that the episode count is present in each title column.
```

```python
def extract_episodes(txt):   # fuction to extract episodes
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True: 
            data = data + i 
        if i == '(':
            check = True 
```

```python
df['Title'].apply(extract_episodes)   # here we extract episodes
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)  # we create a new column name Episodes in dataframe.
```

```python
df['Episodes'] = df['Episodes'].str.replace('eps', "")  # here we replaces string eps with whitespace so that we can perform aggregation functions easily.
```

```python
df['Episodes'] = df['Episodes'].astype(int)    # we convert the str datatypes of episodes number into integers so that we can apply aggregation functions easily.
```

```python
df.head()
```

-make a new column for time stamp

```python
df.loc[0]['Title']  # here we observed that the time stamp is present in each title column.
```

```python

def extraction_time(txt):   # function to extract time stamp
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]
            return data        
```

```python
df['Title'].apply(extraction_time)   # here we extract time stamp
```

```python
df['Time Stamp'] = df['Title'].apply(extraction_time)  # we create a new column name Time Stamp in dataframe.
```

```python
df.head()
```

```python
# function to convert time stamp into number of months

from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None
```

```python
df['Months'] = df['Time Stamp'].apply(calculate_total_months)
```

```python
df.head()
```

```python
# which anime has the highest score

df[df['Score'] == df['Score'].max()]['Title']
```

```python
# give me top 5 highest scoring anime

df['Title'].head()     # dataset is ranked on the basis of score so we can find directly 
```

```python
# which anime has the highest episode count

df[df['Episodes'] == df['Episodes'].max()]['Title']
```

```python
# animes with top 5 episode count

df['Episodes'].value_counts().head(5)
```

```python
#which is the longest running anime

df[df['Months'] == df['Months'].max()]['Title']
```
