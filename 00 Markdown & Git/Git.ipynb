### PUSH A CODE ON GITHUB

- Push your code from local system to remote repositories.

- git --> check if git is working or not<br>
- git init --> use to initialize the git<br>
- git add . OR git add filename --> add all the files for tracking<br>
- git commit -m "any message"<br>
    - if have any error:-<br>
        git config --global user.email "you@example.com"<br>
        git config --global user.name "your username"<br>  
- git branch -m main<br>    
- git branch --> to check the branch<br>
- git remote add origin "link here" <br>
    - if have any error:-<br>
        git remote set-url origin "link here"<br>
- git push -u origin main

### GIT CLONE URL

- Bring files from remote repositories to local system.

- git clone "link here" <br>
- cd filename/ <br>
- git branch<br>
- now suppose we change something in the file we cloned <br>
    - git status --> tells which file is modified.<br>
- now after modification, we have to push the code again on git or github<br>
    - ~~git init~~<br>
    - git add . <br>
    - git commit -m "message for changes" <br>
    - ~~git remote add origin "link here"~~ <br>
    - git push -u origin main/ git push<br>
- now suppose we add new file in the file we cloned <br>
    - git status --> tells file is added or not tracked<br>
- now after modification, we have to push the files again on git<br>
    - git add . <br>
    - git commit -m "message for new file added"<br>
    - git push -u origin main

The number of files in the local repository must be equal to remote repository<br>

- git pull --> this will check from the remote repository vs the current/local repository whether the files are updated or not. If not it will bring the changes of the remote repository into your local repository. Then our local is same as remote repository. The local changes after this not the remote<br>
- git add . OR git add . && git commit -m "update message here" <br>
- git push

### CREATING BRANCHES IN GIT

- git pull<br>
- git branch branch_name (eg test_1)(br)
- git branch --> to check branches<br>
- git checkout branch_name --> to switch the branch<Br>
- git commit -m "new copy added message or anything"<br>
- git push -u origin branch_name
